```python
#---Vendas interativas com plotly
import pandas as pd
import plotly.express as px
```

```python
#---Base de dados
df = pd.DataFrame({
    "Mês": ["Jan", "Fev", "Mar", "Abr", "Mai", "Jun"],
    "Vendas": [1200, 1800, 1500, 2200, 2500, 3000]
})
```

```python
#Gráfico interativo
fig = px.line(df,
              x="Mês",
              y="Vendas",
              markers=True,
              title="Crescimento das Vendas (Interativo)",
              hover_data=["Vendas"],
              line_shape="spline",
              color_discrete_sequence=["#3A86FF"])
```

```python
#---Mostrar gráfico
fig.update_layout(template="plotly_white")
fig.show()
```

```python
# Salvando grafico em HTML
fig.write_html("crescimento_vendas.html")
print("Gráfico salvo como crescimento_vendas.html")
```
